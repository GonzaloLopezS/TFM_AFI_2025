```python
files = '../CBP_drug_seizures/nationwide-drugs-fy19-fy22.csv' # Generar mascara para obtener todos los csv de la carpeta
drug_seizure_2019_DF = pd.read_csv(files,sep=',') # header = infer
```

```python
drug_seizure_2019_DF = drug_seizure_2019_DF.loc[drug_seizure_2019_DF["FY"] == 2019]
```

Año 2020, 21:

```python
files = '../CBP_drug_seizures/nationwide-drugs-fy20-fy23.csv' # Generar mascara para obtener todos los csv de la carpeta
drug_seizure_2020_DF = pd.read_csv(files,sep=',') # header = infer
drug_seizure_2020_DF = drug_seizure_2020_DF.loc[drug_seizure_2020_DF["FY"] == 2020]
```

```python
files = '../CBP_drug_seizures/nationwide-drugs-fy21-fy24.csv' # Generar mascara para obtener todos los csv de la carpeta
drug_seizure_2021_DF = pd.read_csv(files,sep=',') # header = infer
drug_seizure_2021_DF = drug_seizure_2021_DF.loc[drug_seizure_2021_DF["FY"] == 2021]
```

```python
files = '../CBP_drug_seizures/nationwide-drugs-fy22-fy25-dec.csv' # Generar mascara para obtener todos los csv de la carpeta
drug_seizure_2022_25_DF = pd.read_csv(files,sep=',') # header = infer
```

```python
drug_seizure_2022_25_DF
```

Concatenar dataframes de 2019 a 2025:

```python
cbp_seizure_dataframe = pd.concat([drug_seizure_2019_DF, drug_seizure_2020_DF, drug_seizure_2021_DF, drug_seizure_2022_25_DF], axis = 0, ignore_index=True, join='outer')
```

```python
month_map = {
    'JAN': 1, 'FEB': 2, 'MAR': 3, 'APR': 4, 'MAY': 5, 'JUN': 6,
    'JUL': 7, 'AUG': 8, 'SEP': 9, 'OCT': 10, 'NOV': 11, 'DEC': 12
}

# Convertir 'Month' a número
cbp_seizure_dataframe['Month_Num'] = cbp_seizure_dataframe['Month (abbv)'].map(month_map)

cbp_seizure_dataframe['FY'] = cbp_seizure_dataframe['FY'].replace({'2025 (FYTD)': '2025'})
cbp_seizure_dataframe['FY'] = pd.to_numeric(cbp_seizure_dataframe['FY'], downcast='integer', errors='coerce')
```

```python
# Ajustar el año fiscal (suponiendo que el FY comienza en OCT)
cbp_seizure_dataframe['Year'] = cbp_seizure_dataframe['FY']
cbp_seizure_dataframe.loc[cbp_seizure_dataframe['Month_Num'] >= 10, 'Year'] = cbp_seizure_dataframe['FY'] - 1

cbp_seizure_dataframe.rename(columns={'Month_Num':'Month'}, inplace=True)
```

```python
cbp_seizure_dataframe['Year'].unique()
```

```python
# Crear la nueva columna con formato YYYY-MM
cbp_seizure_dataframe['Date'] = pd.to_datetime(cbp_seizure_dataframe[['Year', 'Month']].assign(day=1)).dt.strftime('%Y-%m')
```

Añadir el identificador de cluster del hub portuario, segun diccionario para luego agregarlo en los datos de tratamiento

```python
cbp_seizure_dataframe['Sum Qty (kg)'] = cbp_seizure_dataframe['Sum Qty (lbs)']/2.20462
```
