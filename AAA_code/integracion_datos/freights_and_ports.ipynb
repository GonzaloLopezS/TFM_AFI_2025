Con el diccionario 'dict_ports_freights' ahora se podrá realizar las operaciones de sum() y el left join con ports_df

```python
def add_column_from_reversed_dict(df, column_a, original_dict, new_column_name):
    """Adds a new column using a reversed dictionary, handling lists as values and skipping errors.

    Args:
        df: The Pandas DataFrame.
        column_a: The name of the column whose values are *values* in the original dictionary.
        original_dict: The original dictionary (can have lists as values).
        new_column_name: The name of the new column.

    Returns:
        The modified Pandas DataFrame.
    """

    reversed_dict = {}
    for key, value in original_dict.items():
        if isinstance(value, list):  # Handle lists of values
            for v in value:
                reversed_dict[v] = key  # Map each element in the list to the key
        else:
            reversed_dict[value] = key # Handle single values

    df[new_column_name] = df[column_a].apply(lambda x: reversed_dict.get(x)) # Use .get() to handle missing keys
    return df
```

```python
# Añadir columna "port_name" en freights_df
freight_df_modified = add_column_from_reversed_dict(freight_df, 'PORT NAME', dict_ports_freights, 'port_name')
```

```python
# Columns to aggregate
columns_to_aggregate = ['Loaded InBound DOMESTIC', 'Empty InBound DOMESTIC',
                       'Loaded OutBound DOMESTIC', 'Empty OutBound DOMESTIC',
                       'Total DOMESTIC', 'Loaded InBound FOREIGN',
                       'Loaded OutBound  FOREIGN', 'Total OutBound  FOREIGN',
                       'Total Loaded Grand']

# Group by 'Year' and 'port_name' and sum the specified columns
aggregated_df = freight_df_modified.groupby(['Year', 'port_name'])[columns_to_aggregate].sum().reset_index()
```

```python
# Left join (ports_df, freight_df) on ='port_name' y realizar sum () para columnas en freight_df aplicadas en ports_df
# Group and aggregate freight data
merged_df = pd.merge(ports_df, aggregated_df, on='port_name', how='left')
```
